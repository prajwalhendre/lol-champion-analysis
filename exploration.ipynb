```python
import requests

response = requests.get("https://ddragon.leagueoflegends.com/api/versions.json")

versions = response.json()

latest_version = versions[0]

print(latest_version)
```

```python
url = f"https://ddragon.leagueoflegends.com/cdn/{latest_version}/data/en_US/champion.json"

response = requests.get(url)
champion_data = response.json()

print(champion_data.keys())
```

```python
champions = champion_data["data"]

print(len(champions))

first_champ = list(champions.keys())[0]
print(first_champ)

print(champions[first_champ])
```

```python
response = requests.get("http://cdn.merakianalytics.com/riot/lol/resources/latest/en-US/champions.json")

release_date_data = response.json()

first_5 = list(release_date_data.keys())[:5]

for champion in first_5:
    print(champion, release_date_data[champion])
    print("----")
```

```python
import pandas as pd

rows = []

for champ_name, champ_data in release_date_data.items():
    stats = champ_data["stats"]
    row = {
        "name": champ_data["name"],
        "release_date": champ_data["releaseDate"],
        "resource": champ_data["resource"],
        "attack_type": champ_data["attackType"],
        "roles": champ_data["roles"],
        "positions": champ_data["positions"],
        "hp": stats["health"]["flat"],
        "hp_per_level": stats["health"]["perLevel"],
        "armor": stats["armor"]["flat"],
        "armor_per_level": stats["armor"]["perLevel"],
        "attack_damage": stats["attackDamage"]["flat"],
        "attack_damage_per_level": stats["attackDamage"]["perLevel"],
        "magic_resist": stats["magicResistance"]["flat"],
        "magic_resist_per_level": stats["magicResistance"]['perLevel'],
        "move_speed": stats["movespeed"]["flat"],
        "attack_range": stats["attackRange"]["flat"]
    }

    rows.append(row)

df = pd.DataFrame(rows)

print(df.shape)
print(df.head())
```

```python
ddragon_names = set(champion_data["data"].keys())

meraki_names = set(df["name"].values)

missing = ddragon_names - meraki_names

print(missing)
```

```python
#zaahen not available in most recent data

for name in df["name"].values:
    if "za" in name.lower() or "zaa" in name.lower():
        print(name)
```

```python
df["release_date"] = pd.to_datetime(df["release_date"])

df["release_year"] = df["release_date"].dt.year

print(df[["name", "release_date", "release_year"]].head(10))
```

```python
df= df.reset_index(drop=True)
df["id"] = df.index + 1

champions_table = df.drop(columns=["roles", "positions"])

print(champions_table.head())
```

```python
champion_roles_table = df[["id", "roles"]].explode("roles")
champion_roles_table.columns = ["champion_id", "role"]
champion_roles_table = champion_roles_table.reset_index(drop=True)

print(champion_roles_table.head(10))
```

```python
champion_positions_table = df[["id", "positions"]].explode("positions")
champion_positions_table.columns = ["champion_id", "position"]
champion_positions_table = champion_positions_table.reset_index(drop=True)

print(champion_positions_table.head(10))
```

```python
import sqlite3

conn = sqlite3.connect("lol_champions.db")

champions_table.to_sql("champions", conn, if_exists="replace", index=False)
champion_roles_table.to_sql("champion_roles", conn, if_exists="replace", index=False)
champion_positions_table.to_sql("champion_positions", conn, if_exists="replace", index=False)

print("loaded successfully")

conn.close()
```
